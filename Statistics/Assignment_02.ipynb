```python
# Q.14)Write a Pandas program to create a plot of adjusted closing prices, 
# thirty days and forty days simple moving average of Alphabet Inc. 
# between two specific dates.
s_date=pd.to_datetime('2020-4-1')
e_date=pd.to_datetime('2020-9-30')
df['Date'] = pd.to_datetime(df['Date'])
n_df=(df['Date']>=s_date) & (df['Date']<=e_date)
df1=df.loc[n_df]
df1.set_index('Date',inplace=True)
df1['SMA_30_days']=df1.iloc[:,4].rolling(window=30).mean()
df1['SMA_40_days']=df1.iloc[:,4].rolling(window=40).mean()
plt.figure(figsize=[15,9])
plt.plot(df1['Adj Close'],color='b',label="Adjusted closing price")
plt.plot(df1.SMA_30_days,color='r',label="SMA 30 days")
plt.plot(df1.SMA_40_days,color='g',label="SMA 40 days")
plt.legend()
plt.title("Stock Prices Forecast")
plt.show()
```

```python
#Q.16) Write a Pandas program to create a plot of adjusted closing prices, 
# 30 days simple moving average and 
# exponential moving average of Alphabet Inc. between two specific dates. 
s_date=pd.to_datetime('2020-4-1')
e_date=pd.to_datetime('2020-9-30')
df['Date'] = pd.to_datetime(df['Date'])
n_df=(df['Date']>=s_date) & (df['Date']<=e_date)
df1=df.loc[n_df]
df1.set_index('Date',inplace=True)
df1['SMA_30_days']=df1.iloc[:,4].rolling(window=30).mean()
df1['EMA']=df1.iloc[:,4].ewm(span=10).mean()
plt.figure(figsize=[15,9])
plt.plot(df1['Adj Close'],color='b',label="Adjusted closing price")
plt.plot(df1.SMA_30_days,color='r',label="SMA 30 days")
plt.plot(df1.EMA,color='g',label="EMA")
plt.legend()
plt.title("Stock Prices Forecast")
plt.show()
```

```python
# Q.17)Write a Pandas program to create a plot to visualize daily percentage 
# returns of Alphabet Inc. 
# stock price between two specific dates. 
s_date=pd.to_datetime('2020-4-1')
e_date=pd.to_datetime('2020-9-30')
df['Date'] = pd.to_datetime(df['Date'])
n_df=(df['Date']>=s_date) & (df['Date']<=e_date)
df1=df.loc[n_df]
df2=df1[['Date','Adj Close']]
df2.set_index('Date',inplace=True)
df3=df2.pct_change(periods=1)
plt.figure(figsize=[12,9])
df3['Adj Close'].plot(marker='o',label='Adjacent Close')
plt.legend()
plt.grid()
plt.title("daily percentage returns of Alphabet Inc. stock price")
plt.show()
```

```python
# Q.18) Write a Pandas program to plot the volatility over a period of time of 
# Alphabet Inc. 
# stock price between two specific dates. 

start_date = pd.to_datetime('2020-4-1')
end_date = pd.to_datetime('2020-9-24')                         
df['Date'] = pd.to_datetime(df['Date']) 
new_df = (df['Date']>= start_date) & (df['Date']<= end_date)
df1 = df.loc[new_df]
df2 = df1[['Date', 'Close']]
df3 = df2.set_index('Date')
data_filled = df3.asfreq('D', method='ffill')
data_returns = data_filled.pct_change()
data_std = data_returns.rolling(window=30, min_periods=30).std()
plt.figure(figsize=(20,20))
data_std.plot();
plt.suptitle('Volatility: ', fontsize=12, color='black')
plt.grid(True)
plt.show()
```

```python
#Q.19) Write a Pandas program to create a histogram to visualize daily return distribution of Alphabet Inc. 
# stock price between two specific dates. 

import matplotlib.pyplot as plt
import seaborn as sns
start_date = pd.to_datetime('2020-4-1')
end_date = pd.to_datetime('2020-9-06')                         
df['Date'] = pd.to_datetime(df['Date']) 
new_df = (df['Date']>= start_date) & (df['Date']<= end_date)
df1 = df.loc[new_df]
df2 = df1[['Date', 'Adj Close']]
df3 = df2.set_index('Date')
daily_changes = df3.pct_change(periods=1)
sns.distplot(daily_changes['Adj Close'].dropna(),bins=100,color='purple')
plt.suptitle('Daily percent return of Alphabet Inc', fontsize=12, color='black')
plt.grid(True)
plt.show()
```
